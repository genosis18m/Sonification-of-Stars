```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the dataset from the .npy file
data = np.load('images.npy')
```

```python
data.shape
```

```python
# data is a 4D array of shape (n_images, height, width, channels)
image = data[6]
image.shape
```

#### checking the image in RGB or BGR format

```python
# Display the image using OpenCV
cv2.imshow('Galaxy Image', image)  # This will show the BGR image
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Display the image without conversion
plt.imshow(image)
plt.title('Original_image')
plt.axis('off')
plt.show()

# Convert to RGB format (if needed)
rgb_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

# Display the image with proper RGB conversion
plt.imshow(rgb_image)
plt.title('With RGB Conversion')
plt.axis('off')
plt.show()
```

```python
#Image is in RGB format 
```

```python
# Extract pixel data (RGB values)
height, width, channels = image.shape
pixel_data = image.reshape(height * width, channels)
```

```python
# Store pixel data in a CSV file
df = pd.DataFrame(pixel_data, columns=['R', 'G', 'B'])
df.to_csv('galaxy_image_pixel_data.csv', index=False)
print('Pixel data saved to galaxy_image_pixel_data.csv')
```

```python
# Visualization using matplotlib: display the image
plt.figure(figsize=(8, 8))
plt.imshow(rgb_image)
plt.title('Galaxy Image(BGR)')
plt.axis('off')
plt.show()
```

```python
# Visualization using a heatmap of intensity (for grayscale visualization)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8, 8))
plt.imshow(gray_image, cmap='hot', interpolation='nearest')
plt.title('Galaxy Image Intensity Heatmap')
plt.colorbar()
plt.axis('off')
plt.show()
```

```python
# Visualize individual RGB channels (e.g., Red channel)
plt.figure(figsize=(8, 8))
plt.imshow(image[:, :, 0], cmap='Reds')  # Visualize the Red channel
plt.title("Red Channel Intensity")
plt.axis('off')
plt.show()
```
